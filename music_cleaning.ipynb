# Clean the data

***üßπ Data wrangling and cleaning.***

1. Remove unused columns (unamed, key, time_signature, loudness, track_genre)
2. Remove any duplicate songs using the track ID column

```python
# Look into the data
music.head()
```

```python
music.info()
```

```python
# Remove columns we won't look into such as unamed, key, time_signature, loudness
music.drop(['Unnamed: 0', 'key', 'time_signature', 'loudness', 'track_genre'], axis=1, inplace=True)
```

```python
# Check if columns were removed
music.head()
```

Check for duplicate tracks

```python
# Check if there are any duplicates in track_id
music.duplicated(subset='track_id').sum()
```

```python
# Look into which tracks are duplicated
music[music.duplicated(subset='track_id', keep=False)].sort_values('track_id')
```

```python
# Remove duplicate tracks and keep only one
music.drop_duplicates(subset='track_id', keep='first', inplace=True)
```

```python
# Recheck if there are any duplicates in track_id
music.duplicated(subset='track_id').sum()
```
